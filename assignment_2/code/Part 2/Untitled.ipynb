```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("results/experiments.csv")
df_30 = df[df['Book'].str.contains('30')]
df_50 = df[df['Book'].str.contains('50')]
```

```python
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121)
ax1 = sns.lineplot(x="Step", y="Loss", hue="Book", data=df_30)
ax1.set_title("Loss")

ax2 = fig.add_subplot(122)
ax2 = sns.lineplot(x="Step", y="Acc", hue="Book", data=df_30)
ax2.set_title("Accuracy")
```
